```python
files = [x for x in os.listdir('/cellar/users/ramarty/Data/hla_ii/presentation/affinities/oncogenes/') if 'wt' not in x]
all_dfs = []
for i, f in enumerate(files):
    if i % 100 == 0:
        print i
    df = pd.concat([pd.read_csv('/cellar/users/ramarty/Data/hla_ii/presentation/affinities/oncogenes/{0}'.format(f), \
            skiprows=1, index_col=0, sep='\t'),
                    pd.read_csv('/cellar/users/ramarty/Data/hla_ii/presentation/affinities/tsgenes/{0}'.format(f), \
            skiprows=1, index_col=0, sep='\t')])
    df['allele'] = f.split('.')[0]
    all_dfs.append(df[df.Peptide.isin(candidate_peptides)])
df = pd.concat(all_dfs)
```

Separate file for each peptide

```python
for peptide in candidate_peptides:
    df[df.Peptide == peptide].to_csv('/cellar/users/ramarty/Data/hla_ii/generated_data/neoantigens/candidates/{0}.csv'.format(peptide))
```

```python
def get_gene(x):
    if len(x.split('-')) > 2:
        return x.split('-')[1][:2]
    else:
        return x[:2]
df['Gene'] = df['allele'].apply(get_gene)
```
